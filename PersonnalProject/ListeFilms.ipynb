## Fonctions unitaires à chaque site

```python
def joinList(*arg):
    joined = []
    for i in arg:
        joined.append(i)
    return joined
```

```python
def luminor():
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()
    driver.get("https://www.luminor-hoteldeville.com/")
    time.sleep(1)

    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')

    titres = soup.find_all('div',id='films_semaine')
    titresReturn = []

    for i in titres:
        titre = i.find_all('h3')
        for i in titre:
            titresReturn.append(str(i.text))
            
    driver.quit()
    
    return titresReturn
```

## Fonction principale

```python
start = datetime.now()

listeCine =[]
listeTitre =[]
listeURL =[]
indexCine = list(cinemas)
nbFilm = 0

titre1 = [indexCine[0],luminor()]
time.sleep(1)
titre2 = [indexCine[1],pantheon()]
time.sleep(1)
titre3 = [indexCine[2],saintMich()]
time.sleep(1)
titre4 = [indexCine[3],quartierLatin()]
time.sleep(1)
titre5 = [indexCine[4],odeon()]
time.sleep(1)
titre6 = [indexCine[5],lincoln()]
time.sleep(1)
titre7 = [indexCine[6],publicis()]
time.sleep(1)
titre8 = [indexCine[7],balzac()]
time.sleep(1)
titre9 = [indexCine[8],caumartin()]
time.sleep(1)
titre10 = [indexCine[9],maxLinder()]
time.sleep(1)
titre11 = [indexCine[10],louxor()]
time.sleep(1)
titre12 = [indexCine[11],brady()]
time.sleep(1)
titre13 = [indexCine[12],parnassiens()]
time.sleep(1)
titre14 = [indexCine[13],chaplinDenfer()]
time.sleep(1)
titre15 = [indexCine[14],chaplinLambert()]
time.sleep(1)
titre16 = [indexCine[15],cineastes()]
time.sleep(1)


titres = joinList(titre1,titre2,titre3,titre4,titre5,titre6,titre7,titre8,titre9,titre10,titre11,titre12,titre13,titre14,titre15,titre16)

for t in titres:
    if len(t[1])==1:
        listeCine.append(t[0])
        listeTitre.append(t[1][0])
        nbFilm +=1
    else:
        for i in range(len(t[1])):
            listeCine.append(t[0])
            listeTitre.append(t[1][i])
            nbFilm +=1
            
end = datetime.now()
gap = end-start
print(f"Le temps pris pour analyser tous les sites est de : {gap} minutes ")
```

## Affichage des films à l'affiche

```python
indexXp =[]
for i in range(1,nbFilm+1):
    indexXp.append(f"Film {i}")

import pandas as pd

films = pd.DataFrame(list(zip(listeCine,listeTitre)),columns=['Nom du cinéma','Nom du film'],
                  index=indexXp)
films['Nom du film'] = films['Nom du film'].str.title()
films
```
