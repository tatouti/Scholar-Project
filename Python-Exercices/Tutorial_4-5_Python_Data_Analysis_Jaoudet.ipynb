#### 1) Create the same chart as 1.1 using the new method: instantiate figure objects and then call methods or attributes from that object.

```python
fig, ax = plt.subplots()
x = np.arange(1,5,0.5)
y = x**2

ax.plot(x,y,c='red')
ax.set_xlabel("X")
ax.set_ylabel("Y = X**2")
ax.set_title("API Plot")
```

#### 2) Create the figure below (the blue curve represents $y=x^2$ while the red curve represents $x=y^2$)

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.48,0.25,0.4,0.3])

axes1.plot(y,x,c='red',marker='s')
axes1.set_xlabel("Y original")
axes1.set_ylabel("X original")
axes1.set_title("Big plot")

axes2.plot(x,y,c='yellow',marker='o')
axes2.set_xlabel("X original")
axes2.set_ylabel("Y original")
axes2.set_title("Small plot")
```

Create a dataframe using this data, marking the correlation between each variable in the heatmap. Add a title and change the style of this plot. Here‚Äôs an example of the expected output:

```python
import seaborn as sns
vehicules_2016HM = vehicules_2016.pivot_table(index='label_vehicule',columns='label_obstacle_mobile',values='nb_occupants',aggfunc="sum")
vehicules_2016HM = vehicules_2016HM[vehicules_2016HM['inconnu_0']>0].sort_values(by='label_vehicule',ascending=True)
vehicules_2016HM = vehicules_2016HM.dropna(how='all',axis=1)
hm =sns.heatmap(vehicules_2016HM,cmap="Oranges")
hm.set(title="Nombre d'accident par type de transport et obstacle")
       
```

#### 5) Create two new columns, the first will contain the hour of the accident, the second will contain the minute. Which hour is the most dangerous for drivers?

```python
def heure(x):
    if len(str(x))==4:
        return int(str(x)[0:2])
    else :
        return int(str(x)[0:1])
        
def minute(x):
    if len(str(x))==4:
        return int(str(x)[-2:])
    else :
        return int(str(x)[-2:])     
```

```python
caracteristiques_2016['hour'] = caracteristiques_2016['hrmn'].map(heure)
caracteristiques_2016['minute'] = caracteristiques_2016['hrmn'].map(minute)
caracteristiques_2016.sort_values(by='hrmn',ascending=True)
```

```python
caractH = caracteristiques_2016.groupby(['hour','minute']).count().sort_values(by='Num_Acc',ascending=False).head(4)
caractH
```

The most dangerous hour for accidents is at 6 pm. We can understand this number because it is usually the time at which people leave their work to go home. So more people on road and so more chances of accidents.

#### 8) Create a new column representing 4 period of the day. Morning, afternoon, evening, night. Use this new column to analyze any correlations with other columns. Present your results in one or more graphs.

```python
def period(x):
    if 0<=x<4:
        return str('Night')
    elif 4<x<=12:
        return str('Morning')
    elif 12<x<=16:
        return str('Afternoon')
    elif 16<x<=20:
        return str('Evening')
    elif 20<x<24:
        return str('Night')
```

```python
caracteristiques_2016['Period'] = caracteristiques_2016['hour'].map(period)
caracteristiques_2016
```

```python
caractP = caracteristiques_2016.groupby('Period').count()
caractP.plot(kind='pie',y='Num_Acc',figsize=(12,7),autopct='%1.0f%%')
```

We can see that the moment of the day with the most accidents is in the morning. This is interesting because most of the accidents happen at 6 pm as we saw earlier. But if we do a more global study with the period of the day, we have morning on top. It is understandable because we have a peek between 5 and 7 pm, but a hollow for the rest of the evening. And it is also logical that we have less accidents during the night because people are sleeping, so less cars on the road.
